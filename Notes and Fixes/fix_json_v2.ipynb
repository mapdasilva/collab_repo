```python
#GCOSME
import json
import os
import glob
jsons=glob.glob(os.path.join(os.path.dirname(os.getcwd()),'BSCMR_BIDS_ISC','**','*_task-ISC_bold.json'), recursive=True)
slicetiming_json=os.path.join(os.getcwd(),'slicetiming.json')
with open(slicetiming_json, 'r') as file:
    data = json.load(file)
    slice_timing=data["SliceTiming"]
```

```python
for json_file in jsons:
    
    # Read the existing JSON file
    with open(json_file, 'r') as file:
        data = json.load(file)
        data["SliceTiming"] = slice_timing
        
        with open(json_file, 'w') as file:
            json.dump(data, file, indent=4)
            print(json_file.split(os.sep)[5])
```
